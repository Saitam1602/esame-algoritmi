```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.read_csv("../data/processed/out.csv")
df.head(1)
```

```python
X_columns = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF',
       'FullBath', 'YearBuilt', '1stFlrSF', 'YearRemodAdd', 'GarageYrBlt',
       'TotRmsAbvGrd']

y_columns = ["SalePrice"]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42)

param_grid = {'max_depth': [None] + [x for x in range(0, 20, 5)], "min_samples_leaf": [
    None] + [x for x in range(0, 20, 5)], "min_samples_split": [None] + [x for x in range(0, 20, 5)]}


grid_search = GridSearchCV(model, param_grid=param_grid, verbose=0, cv=5)
grid_search.fit(df[X_columns], df[y_columns])

grid_search.best_params_

best_params = {'max_depth': 8, 'min_samples_leaf': 2,'min_samples_split': 8}

# max_depth: 8 {'min_samples_leaf': 2} {'min_samples_split': 8}

```

```python
best_params = {'max_depth': 8, 'min_samples_leaf': 2,'min_samples_split': 8}

model = RandomForestRegressor(random_state=42, **best_params)

model.fit(df[X_columns], df[y_columns])
```

```python
oe = OrdinalEncoder()

df_test = pd.read_csv("../data/raw/test.csv")

categorical_columns = df_test.select_dtypes(include=['object']).columns.tolist()

for c in categorical_columns:
    df_test[[c]] = oe.fit_transform(df_test[[c]])

```
