````python
import pandas as pd
import pickle

from data.make_dataset import load_data
df = load_data("../data/raw/train.csv")
df.head()
````

````python
from features.build_features import encode_categorical_data, get_best_correlation_columns

df = encode_categorical_data(df)
df
````

````python
best_correlation_column = get_best_correlation_columns(df, 'SalePrice', .5)
best_correlation_column
````

````python
from features.build_features import choose_columns
X_train = choose_columns(df, best_correlation_column)

y_train = choose_columns(df, ['SalePrice'])
````

````python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
````

## Decision Tree Best Params (GridSeach):

```python
{'criterion': 'absolute_error',
 'max_depth': 8,

```

````python
decision_tree_regressor_model_best_params = {
    "criterion": "absolute_error",
    "max_depth": 8,
    "min_samples_leaf": 2,
    "min_samples_split": 18,
}

decision_tree_regressor_model = DecisionTreeRegressor(**decision_tree_regressor_model_best_params)
decision_tree_regressor_model.fit(X_train, y_train)

filename = 'models/decision_tree_regressor_model.sav'
pickle.dump(decision_tree_regressor_model, open(filename, 'wb'))
````

# Random Forest Regressor

````python
model = RandomForestRegressor(n_estimators=50)

param_grid = {
    'criterion': ["squared_error", "absolute_error"],
    'max_depth': [None] + [x for x in range(0,20,5)],
    'min_samples_leaf': [None] + [x for x in range(0,20,5)],
    'min_samples_split': [None] + [x for x in  range(0,20,5)]
}

grid_search = GridSearchCV(model, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

grid_search.best_params_
````
